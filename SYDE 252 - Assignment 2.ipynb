```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets 
import numpy as np
import matplotlib.pyplot as plt

#dropdown menu for the shapes of functions
f_shape = widgets.Dropdown(options=['step','impulse','triangle','pulse','cosine'])
g_shape = widgets.Dropdown(options=['step','impulse','triangle','pulse','cosine'])

#f(t-tau) widgets
#common widget for all shapes
amplitude_f = widgets.FloatSlider(min=-10, max=10, step=0.1, description="Amplitude")
#non-cosine shape widgets
width_f = widgets.FloatSlider(min=0, max=10, step=0.1, description="Width")
displacement_f = widgets.FloatSlider(min=-10, max=10, step=0.1, description="Displacement")
#cosine shape widgets
frequency_f = widgets.FloatSlider(min=0, max=10, step=0.1, description="Frequency")
phase_f = widgets.FloatSlider(min=-10, max=10, step=0.1, description="Phase")

#g(tau) widgets
#common widget for all shapes
amplitude_g = widgets.FloatSlider(min=-10, max=10, step=0.1, description="Amplitude")
#non-cosine shape widgets
width_g = widgets.FloatSlider(min=0, max=10, step=0.1, description="Width")
displacement_g = widgets.FloatSlider(min=-10, max=10, step=0.1, description="Displacement")
#cosine shape widgets
frequency_g = widgets.FloatSlider(min=0, max=10, step=0.1, description="Frequency")
phase_g = widgets.FloatSlider(min=-10, max=10, step=0.1, description="Phase")

#allows the selection of Continuous or Discrete Domain
domain_selector = widgets.Checkbox(description='Discrete')

#Button to update the graph
plot_button = widgets.ToggleButton(description="Plot the Functions")

#variables that hold the functions f and g for plotting
f_function = 0
g_function = 0

#Main Function for UI Generation and Graphing-----------------------------------
def main(f, g, plot, domain):
    #range of the graphs dependent on cts or discrete domains
    x_val = np.arange(-20.0, 20.0, 0.01)
    if(domain):
        x_val = np.arange(-20.0, 20.0, 1.0)
    #shows the UI for function F and adjusts depending of shape selection
    print('Controls for Function F: ')
    if(f == 'cosine'):
        #displays the widgets used to select values
        display(amplitude_f)
        display(frequency_f)
        display(phase_f)
        #sets the f function to be equal to the cosine shape with given params from sliders
        f_function = cosine(amplitude_f.value, frequency_f.value, phase_f.value, x_val)
        #will graph with dots or no dots depending on domain
        if(domain):
            plt.plot(x_val, (f_function), 'ro', color="blue")
        else:
            plt.plot(x_val, (f_function), color="blue")
    #functionally analogous to previous
    elif(f == 'impulse' or f == 'step'):
        display(amplitude_f)
        display(displacement_f)
        if(f == 'step'):
            f_function = step(amplitude_f.value, displacement_f.value, x_val)
            if(domain):
                plt.plot(x_val, (f_function), 'ro', color="blue")
            else:
                plt.plot(x_val, (f_function), color="blue")
        elif(f == 'impulse'):
            f_function = impulse(amplitude_f.value, displacement_f.value, x_val)
            if(domain):
                plt.plot(x_val, (f_function), 'ro', color="blue")
            else:
                plt.plot(x_val, (f_function), color="blue")
    #functionally analogous to previous
    else:
        display(amplitude_f)
        display(width_f)
        display(displacement_f)
        if(f == 'pulse'):
            f_function = pulse(amplitude_f.value, width_f.value, displacement_f.value, x_val)
            if(domain):
                plt.plot(x_val, (f_function), 'ro', color="blue")
            else:
                plt.plot(x_val, (f_function), color="blue")
        elif(f == 'triangle'):
            f_function = triangle(amplitude_f.value, width_f.value, displacement_f.value, x_val)
            if(domain):
                plt.plot(x_val, (f_function), 'ro', color="blue")
            else:
                plt.plot(x_val, (f_function), color="blue")
    #Repeats all the same logic but for the g function this time all code is functionally analagous to f code
    print('Controls for Function G: ')
    if(g == 'cosine'):
        display(amplitude_g)
        display(frequency_g)
        display(phase_g)
        g_function = cosine(amplitude_g.value, frequency_g.value, phase_g.value, x_val)
        if(domain):
            plt.plot(x_val, (g_function), 'ro', color="red")
        else:
            plt.plot(x_val, (g_function), color="red")
    elif(g == 'impulse' or g == 'step'):
        display(amplitude_g)
        display(displacement_g)
        if(g == 'step'):
            g_function = step(amplitude_g.value, displacement_g.value, x_val)
            if(domain):
                plt.plot(x_val, (g_function), 'ro', color="red")
            else:
                plt.plot(x_val, (g_function), color="red")
        elif(g == 'impulse'):
            g_function = impulse(amplitude_g.value, displacement_g.value, x_val)
            if(domain):
                plt.plot(x_val, (g_function), 'ro', color="red")
            else:
                plt.plot(x_val, (g_function), color="red")
    else:
        display(amplitude_g)
        display(width_g)
        display(displacement_g)
        if(g == 'pulse'):
            g_function = pulse(amplitude_g.value, width_g.value, displacement_g.value, x_val)
            if(domain):
                plt.plot(x_val, (g_function), 'ro', color="red")
            else:
                plt.plot(x_val, (g_function), color="red")
        elif(g == 'triangle'):
            g_function = triangle(amplitude_g.value, width_g.value, displacement_g.value, x_val)
            if(domain):
                plt.plot(x_val, (g_function), 'ro', color="red")
            else:
                plt.plot(x_val, (g_function), color="red")
    
    #multiplies the functions together to give the third function needed
    f_multi_g = f_function*g_function
    if(domain):
        plt.plot(x_val, (f_multi_g), 'ro', color="yellow")
    else:
        plt.plot(x_val, (f_multi_g), color="yellow")
    #adds the additionally labels to the graph before plotting the first graph
    plt.title('f(t-tau), g(tau), and f(t-tau)g(tau)')
    plt.legend(['f(t-tau)', 'g(tau)' ,'f(t-tau)g(tau)'])
    plt.xlabel('tau')
    plt.ylabel('Values')
    #showing the graph ends the first graph all code after will make a new graph
    plt.show()
    
    #function call to graph covolution graph
    convolution(f_function, g_function, x_val, domain)
    
    #sets the plot button value to false so the user can now click again as all processing is complete
    #allows the UI to run smoothly for the user at the cost of real time updates (laggy)
    plot_button.value = False
#----------------------------------------------
        
#mathematical definition of each function------
#defined as piecewise functions depending on the range x returning a function that when ploted will be in the function named shape
def step(amp, disp, x):
    return amp*(np.piecewise(x, [x < disp, x >= disp] , [0, 1]))

def impulse(amp, disp, x):
    return (np.piecewise(x, [x < disp, (np.abs(x - disp) <= 0.01), x > disp] , [0, 1, 0]))

def triangle(amp, wth, disp, x):
    return np.piecewise(x, [x - disp < -wth, (x - disp > -wth) & (x - disp < 0), (x - disp > 0) & (x - disp < wth), x - disp > wth], [0, lambda x: ((amp/wth)*x) + (amp - ((amp/wth)*disp)), lambda x: ((-amp/wth)*x) + (amp + ((amp/wth)*disp)), 0])

def pulse(amp, wth, disp, x):
    return (np.piecewise(x, [x < (-(wth/2))+disp, (x >= (-(wth/2))+disp) & (x < (wth/2)+disp), x >= (wth/2)+disp] , [0, amp, 0]))

def cosine(amp, freq, phs, x):
    return amp*(np.cos((freq*x)-phs))

#-----------------------------------------------

#Convolution Calculations-----------------------

def convolution(f_function, g_function, x, domain_val):
    #uses convolve function to get a function of the convolution of f and g
    convolved = np.convolve(f_function, g_function, mode='same')
    #plots a cts or discrete graph depending on the domain set by user
    if(domain_val):
        plt.plot(x, (convolved), 'ro', color="red")
    else:
        plt.plot(x, (convolved), color="red")
    #sets the titles for the graph
    plt.title('Convolution of f(t-tau) & g(tau)')
    plt.xlabel('tau')
    plt.ylabel('Values')

#-----------------------------------------------

#title for the whole program
print('Continuous Linear Convolution')
#allows the graph to be interactive and update passed on the widgets inputed as parameters
interact(main, f=f_shape, g=g_shape, plot=plot_button, domain=domain_selector)
```
